```python
def sum_python(n):
  numbers = range(0, n)
  acc = 0
  for i in numbers:
    acc = acc + i
  return acc
```

```python
import numpy as np
def sum_numpy(n):
  numbers = np.arange(0, n, dtype = np.int64)
  return numbers.sum()
```

```python
def sum_sq_python(n):
  numbers = range(0, n)
  acc = 0
  for i in numbers:
    acc = acc + i**2
  return acc

def sum_sq_numpy(n):
  numbers = np.arange(0, n, dtype = np.int64)
  return np.square(numbers).sum()

from tinybench import benchmark, benchmark_env
bench = benchmark(['sum_sq_python(10 ** 6)', 'sum_sq_numpy(10 ** 6)'], ntimes = 100, warmup = 10, g = globals())
bench.plot()  
```

```python
bench.means['sum_sq_python'] / bench.means['sum_sq_numpy']
```

```python
def var_python(n):
  numbers = range(0, n)
  mean = sum(numbers) / n
  return sum([(x - mean) ** 2 for x in numbers]) / (n - 1)

def var_numpy(n):
  numbers = np.arange(0, n, dtype = np.float64)
  return numbers.var(ddof = 1)

var_numpy(10**5)
var_python(10**5)
```
